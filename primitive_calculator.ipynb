```python
def optimal_sequence(num):
    # create dp table
    dp = [0] * (num+1)
    for i in range(0, num+1):
        if i == 0:
            dp[i] = 0
        elif i == 1:
            dp[i] = 1
        elif i == 2:
            dp[i] = 1
        elif i == 3:
            dp[i] = 1
            
        elif (i%3 == 0) and (i%2 == 0):
            
            d3 = 1 + dp[i//3]
            d2 = 1 + dp[i//2]
            a1 = 1 + dp[i-1]
            
            m = min(a1, d2, d3)
            dp[i] = m
                
        elif (i%3 == 0):
            
            d3 = 1 + dp[i//3]
            a1 = 1 + dp[i-1]
            
            m = min(a1, d3)
            dp[i] = m
 
        elif (i%2 == 0):
            
            d2 = 1 + dp[i//2]
            a1 = 1 + dp[i-1]
            
            m = min(a1, d2)
            dp[i] = m
            
        else:
            dp[i] = dp[i-1] + 1

    # reconstruct the sequence by backtracking
    seq = [num]
    while num > 1:
        if num <= 3:
            seq.append(1)
            break
        elif (num%3 == 0) and (dp[num//3] == dp[num]-1):
                seq.append(num//3)
                num = num // 3
        elif (num%2 == 0) and (dp[num//2] == dp[num]-1):
                seq.append(num//2)
                num = num // 2
        else:
            seq.append(num-1)
            num = num - 1
    
    return dp, seq[::-1]
```

```python
dp, seq = optimal_sequence(96234)
```
