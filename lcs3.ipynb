```python
import numpy as np
def lcs3(a, b, c):
    m = len(a)+1
    n = len(b)+1
    l = len(c)+1
    
    dp = np.zeros((m, n, l))
    
    for i in range(m):
        for j in range(n):
            for k in range(l):
                if i==0 or j==0 or k==0:
                    dp[i][j][k] = 0
                    
                elif a[i-1]==b[j-1] and b[j-1]==c[k-1]:
                    dp[i][j][k] = dp[i-1][j-1][k-1] + 1
                    
                else:  
                    dp[i][j][k] = max( dp[i][j-1][k],     
                                       dp[i][j][k-1],     
                                       dp[i-1][j][k])     
        
    common_sequence = []
    i = m-1
    j = n-1
    k = l-1
    while i>0 and j>0 and k>0:
        if a[i-1]==b[j-1] and b[j-1]==c[k-1]:
            common_sequence.append(a[i-1])
            i -= 1
            j -= 1
            k -= 1
        
        # go to the biggest dp position
        elif dp[i-1][j][k] > dp[i][j-1][k]:
            if dp[i-1][j][k] > dp[i][j][k-1]:   
                i -= 1
            else:
                k -= 1
        else:
            if dp[i][j-1][k] > dp[i][j][k-1]:
                j -= 1
            else:
                k -= 1

    return dp, common_sequence
```

```python
lcs3([1,2,3], [2,1,3], [1,3,5])
```

```python
lcs3([8,3,2,1,7], [8,2,1,3,8,10,7], [6,8,3,1,4,7])
```
