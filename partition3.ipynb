import numpy as np
def distribute_to_3(lst, mean):
    if int(mean) != mean:
        return -1
    else:
        mean = int(mean)
    
    m = len(lst) + 1
    n = mean + 1
    
    dp = np.zeros((2, n))
    
    for i in range(1, m):
        r = i%2
        for j in range(1, n):
            num = lst[i-1]
            if r == 0:
                if num > j:
                    dp[r][j] = dp[r+1][j]
                else:
                    dp[r][j] = max( num + dp[r+1][j-num],
                                    dp[r+1][j])
            else:
                if num > j:
                    dp[r][j] = dp[r-1][j]
                else:
                    dp[r][j] = max( num + dp[r-1][j-num],
                                    dp[r-1][j])
                    
                    
                    
    if r == 0:
        dp[0], dp[1] = dp[1], dp[0]
    return dp

```python
def precheck(lst):
    mean = sum(lst)/3
    if int(mean) != mean or len(lst) < 3:
        return False
    else:
        return True
```

```python
import numpy as np
def distribute_to_3pro(lst, mean):
    m = len(lst) + 1
    n = mean + 1
    
    dp = np.zeros((m, n))
    lst = sorted(lst)
    for i in range(1, m):
        for j in range(1, n):
            num = lst[i-1]
            if num > j:
                dp[i][j] = dp[i-1][j]
            else:
                dp[i][j] = max( num + dp[i-1][j-num],
                                      dp[i-1][j])
                
            
    lst_copy = lst[:]
    i = m-1
    j = mean
    while j > 0 and i > 1:
        num = lst[i-1]
        if (num + dp[i-1][j-num]) >= dp[i-1][j]:
            if num > j:
                i -= 1
            else:
                lst_copy.remove(num)
                i -= 1
                j -= num
        else:
            i -= 1
    return dp, lst_copy
```

```python
def main():
    lst = list(map(int, input().strip().split()))
    #lst = [17, 59, 34, 57, 17, 23, 67, 1, 18, 2, 59]
    mean = sum(lst)/3
    
    if precheck(lst):
        mean = int(mean)
        dp, lst = distribute_to_3pro(lst, mean)
    else:
        return 0
    
    dp, lst = distribute_to_3pro(lst, mean)
    if dp[-1][-1] == mean:
        return 1
    else:
        return 0
```

```python
main()
```
